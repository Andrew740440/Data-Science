```python
import pandas as pd
```

```python
import numpy as np
```

```python
url = 'Corona Virus Dataset.csv'
```

```python
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.rename(columns = {'Country/Region':'Country'},inplace =True)
```

```python
df.head()
```

```python
df = df[['Country','Confirmed','Deaths','Recovered']]
```

```python
df.head()
```

```python
df = df.groupby(['Country'],as_index =False).sum()
```

```python
df.set_index('Country',inplace =True)
```

```python
df.value_counts()
```

```python
df.head()
```

```python
# Find top 5 countries having highest Confirmed case
```

```python
df.sort_values(by = 'Confirmed',ascending = False,inplace = True)
```

```python
df.head()
```

```python
df_confirmed = df['Confirmed']
```

```python
df_confirmed.head()
```

```python
df_confirmed = df_confirmed.head(5)
```

```python
df_confirmed
```

```python
df_confirmed = df_confirmed.transpose()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df_confirmed.plot(kind = 'bar')
plt.xlabel('Country')
plt.ylabel('Number of Confirmed cases')
plt.title('Top 5 Countries with the highest number of Confirmed Cases')

plt.show()
```

```python
# Find Top 5 countries having highest Deaths case
```

```python
df.sort_values(by = 'Deaths',ascending = False,inplace = True)
```

```python
df_Deaths = df['Deaths']
```

```python
df_Top5_Deaths = df_Deaths.head(5)
```

```python
df_Top5_Deaths.plot(kind = 'line',color = 'Red')
plt.xlabel('Number of Deaths')
plt.ylabel('Country')
plt.title('Top 5 Countries with highest number of death')
plt.show()
```

```python
#  Question 7. Find the  frequent distribution of recovered cases
```

```python
import numpy as np
```

```python
df.head()
```

```python
counts,bin_edges = np.histogram(df['Recovered'])
```

```python
counts
```

```python
bin_edges
```

```python
df['Recovered'].plot(kind = 'hist',alpha = 0.25,figsize = (20,10),xticks = bin_edges,color = 'Green')
plt.xlabel('Number of Recovered cases')
plt.ylabel('Number of Countries')
plt.title('Frequency distribution of Number of Recovered cases')
plt.show()
```
